```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("amazon_prime_users.csv")
df.sample(5)
```

# UNDERSTANDING THE DATA

```python
# understanding the data

df.shape
```

```python
df.info()
```

```python
# dropping the unwanted columns

drop_columns=[0,1,2,3,7,8,10]
```

```python
df.drop(df.columns[drop_columns],axis=1,inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
# changing the data type of date of birth

df["Date of Birth"]=pd.to_datetime(df["Date of Birth"])
```

```python
cols=df.columns
color=["#F0FFFF","#069AF3"]
sns.heatmap(df[cols].isnull(),cmap=sns.color_palette(color))
plt.title("NULL VALUES VISUALIZATION",fontsize=12)
```

```python
df[df.duplicated()].shape
```

```python
# calculating the age of the users

from datetime import datetime as dt
```

```python
df["Age"]=dt.today().year-df["Date of Birth"].dt.year
```

```python
df.pop("Date of Birth")
```

```python
df["Age"].min()
```

```python
# categorizing the users as per their age

def label(num):
  if num <=30:
    return "Gen Z"
  elif num >=31 and num <=45:
    return "Millenials"
  elif num >=46 and num <=60:
    return "Gen X"
  elif num >=61 and num <=79:
    return "Boomers"
  else:
    return "Silent Gen"
```

```python
df["label"]=df["Age"].apply(label)
```

# DATA VISUALIZATION

```python
# Understanding the correlation

cor_data=df[["Age","Customer Support Interactions","Feedback/Ratings"]].corr()
sns.heatmap(cor_data,annot=True,cmap="flare")
```

```python
colors=["#FFFF14","#000000","#DC143C","#FF81C0"]
plt.figure(figsize=(15,5))
ax=sns.countplot(x="label",data=df,palette=colors)
plt.tight_layout()
```

```python
# generation vs subscription  plan

colors=["#000000","#C1F80A","#FFD700","#FF81C0","#0000FF"]
sns.set_style(style="darkgrid")
plt.figure(figsize=(15,5))
sns.countplot(x="Subscription Plan",data=df,hue="label",palette=colors)
sns.despine(left=False,right=True,top=True,bottom=False)
plt.legend(loc="upper left")
plt.title("Subscription plan as per generation",fontsize=12)
plt.tight_layout()
```

```python
print(df.groupby("Subscription Plan")["label"].value_counts(normalize=True).round(3)*100)
```

```python
# usage frequency analysis as per generation

plt.figure(figsize=(15,5))
sns.countplot(x="Usage Frequency",data=df,hue="label",palette=colors)
sns.despine(left=False,right=True,top=True,bottom=False)
plt.tight_layout()
```

```python
print(df.groupby("Usage Frequency")["label"].value_counts(normalize=True).round(3)*100)
```

```python
# Genre Analysis
genres=df.groupby("label")["Favorite Genres"].value_counts().sort_values(ascending=False)
sorted_genres=genres.reset_index()
sorted_genres
plt.figure(figsize=(15,7))
sns.barplot(x="count",data=sorted_genres,y="label",hue="Favorite Genres",palette=colors)
sns.despine(left=False,right=True,top=True,bottom=False)
plt.tight_layout()
```

```python
print(df.groupby("Favorite Genres")["label"].value_counts(normalize=True).round(3)*100)
```

```python
colors=["#580F41","#800000"]
plt.figure(figsize=(15,5))
sns.countplot(x="Favorite Genres",data=df,hue="Gender",palette=colors)
sns.despine(left=False,right=True,top=True,bottom=False)
plt.tight_layout()
```

```python
print(df.groupby("Favorite Genres")["Gender"].value_counts(normalize=True).round(3)*100)
```

```python
# user behaviour
plt.figure(figsize=(15,7))
colors=["#FFFF14","#E50000","#808000","#006400","#008080"]
sns.countplot(x="Purchase History",data=df,hue="label",palette=colors)
```

```python
print(df.groupby("Purchase History")["label"].value_counts(normalize=True).round(3)*100)
```

```python
grouped=df.groupby("label")["Customer Support Interactions"].sum()
df_grouped=grouped.reset_index()
plt.figure(figsize=(15,5))
sns.barplot(data=df_grouped,x="label",y="Customer Support Interactions",palette=colors)
sns.despine(left=False, right=True, top=True, bottom=False)
plt.tight_layout()
```
